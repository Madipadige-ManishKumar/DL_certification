```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import logging
logging.getLogger("tensorflow").disabled=True
warnings.simplefilter("ignore")
df = pd.read_csv("weight-height.csv")
df.head()
def plot_human():
    df.plot(kind="scatter",x="Height",y="Weight",c="blue")

plot_human()
# df.plot([78,50],[0,0],c="blue",marker="_")

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam ,SGD

model =Sequential()
model.add(Dense(1,input_shape=(1,)))
model.summary()
```

```python
model.compile(Adam(0.8),'mean_squared_error')
model.fit(df['Height'],df['Weight'],epochs=10,verbose=1)
df.plot(kind='scatter',x='Height',y='Weight')
# plt.plot(df[height])`
x_line = np.linspace(df['Height'].min(), df['Height'].max(), 100)

# Predict weights for the generated heights
y_line = model.predict(x_line)

# Plot the line on the scatter plot
plt.plot(x_line, y_line, color='red', label='Fitted Line')
```

```python
import numpy as np
def line(x,w=0,b=0):
    return (w*x+b)
    pass
def mean_squared_error(y_true,y_pred):
    s=(y_true-y_pred)**2
    return s.mean()
X=df[['Height']].values
y_true=df['Weight'].values
y_pred=line(X,w=2,b=15)
y_pred
mean_squared_error(y_true,y_pred)
# df.plot(kind='scatter',x='Height',y='Weight')
# plt.plot(X,line(X,w=5),linewidth=5)
plt.figure(figsize=(10,5))
ax1=plt.subplot(121)
# df.plot(kind="scatter",x="Height",y="Weight",ax=ax1)
bbs=np.array([-100,-50,0,50,100,150])
result=[]
for b in bbs:
    y_pred=line(X,w=2,b=b)
    result.append(mean_squared_error(y_true,y_pred))
# plt.plot(X,y_pred)
plt.plot(bbs,result,'o-')
plt.xlabel('b')
plt.tight_layout()
print(result)
```
